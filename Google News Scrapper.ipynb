```python
# the_html = driver---somehow----.get_attribute('innerHTML')
# bs = BeautifulSoup(the_html, 'html.parser')
# html = driver.execute_script("return document.body.innerHTML;")
# bs = BeautifulSoup(the_html, 'html.parser')

def getPageContent(url):
    try:
        driver.get(url)
        print(url)
        sleep(3)
        corpus = driver.find_elements_by_tag_name('p')
        scape_text = ''
        for i in corpus:
    #         print(i.text)
            if(i.text != ""):
                content = i.text.strip()
                print('CONTENT:', content)
    #             scape_text.join(content)
                scape_text+= content
        print('SCARP TETX:', scape_text)
    except:
        pass
    
    return scape_text
```

```python
def callURL():
    for news in news_dict:
#         print(news_dict[news])
        try:
            scape_text = getPageContent(news_dict[news])
            print(scape_text)
            with open("corpus/"+news, "a") as file:
                file.write(scape_text)
        except:
            pass
```
